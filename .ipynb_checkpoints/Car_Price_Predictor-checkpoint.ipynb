## Cleaning the Dataset 

#### year has many non-year values

```python
# Filters out the cars with numeric values
car=car[car['year'].str.isnumeric()]
```

```python
car
```

#### year is in object. Change to integer

```python
car['year']=car['year'].astype(int)
```

```python
car.info()
```

####  Price has 'Ask for Price', Remove that.

```python
car=car[car['Price']!="Ask For Price"]
```

```python
car
```

#### Price has commas in its prices and is in object

```python
# Replacing ',' with "" in car['Price'] and converting to integer
car['Price']= car['Price'].str.replace(",","").astype(int)
```

```python
car.info()
```

#### kms_driven has object values with kms in suffix. Chop it off.

```python
car['kms_driven']= car['kms_driven'].str.split(" ").str.get(0).str.replace(",","")
```

```python
car['kms_driven']
```

#### To neglect non numeric values 

```python
car = car[car['kms_driven'].str.isnumeric()]
```

```python
car['kms_driven']=car['kms_driven'].astype(int)
```

```python
car.info()
```

#### Remove nan values from fuel type

```python
car = car[~car['fuel_type'].isna()]
```

```python
car
```

#### **name and company had spammed data , but with the previous cleaning, those rows got removed**
#### Company does not need any cleaning now. Changing car names. Keeping only the first three words

```python
car['name']=car['name'].str.split(" ").str.slice(0,3).str.join(' ')
```

#### Resetting the index of the final cleaned data

```python
car=car.reset_index(drop=True)
```

### Model

```python
x = car.drop(columns="Price")
```

```python
y = car['Price']
```

### Applying Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

#### Creating an OneHotEncoder object to contain all the possible categories

```python
ohe = OneHotEncoder()
ohe.fit(x[['name','company','fuel_type']])
```

#### Creating a column transformer to transform categorical columns

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),
                                    remainder='passthrough')
```

#### linear regression model

```python
lr=LinearRegression()
```

#### Making a pipeline

```python
pipe=make_pipeline(column_trans,lr)
```

#### Fitting the model

```python
pipe.fit(x_train,y_train)
```

#### Checking R2 Score

```python
y_pred=pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

#### Finding the model with a random state of TrainTestSplit where the model was found to give almost 0.92 as r2_score

```python
scores=[]  
for i in range(1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(x_train,y_train)
    y_pred=pipe.predict(x_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

#### The best model is found at a certain random state

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(x_train,y_train)
y_pred=pipe.predict(x_test)
r2_score(y_test,y_pred)
```
